```python
#Total Sales from all items
Total_Sales=df['Sales'].sum()


#Average Sale
Avg_Sale=df['Sales'].mean()

#No Of Item Sold
No_Of_Item_Sold=df['Sales'].count()

#Average Rating
Avg_Rating=df['Rating'].mean()








print(f"Total Sales: ${Total_Sales:,.2f}")
print(f"Average Sale: ${Avg_Sale:,.2f}")
print(f"No Of Item Sold: {No_Of_Item_Sold:}")
print(f"Average Rating: {Avg_Rating:,.2f}")




```

## **Visuals For The Analysis**

### **Sales Based on Fat Content**

```python
Sales_by_fat = df.groupby('Item Fat Content')['Sales'].sum()

plt.pie(Sales_by_fat,
        labels=Sales_by_fat.index,
        autopct='%.2f%%',
        startangle=90)

plt.axis("equal")
plt.title("Sales By Fat Content")
plt.show()
```

### **Total Sales by item type**

```python
Total_Sales_By_Item_Type=df.groupby('Item Type')['Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(13, 5))

bars=plt.bar(Total_Sales_By_Item_Type.index,
        Total_Sales_By_Item_Type.values)


plt.xticks(rotation=90)
plt.title("Sales Item Type")
plt.xlabel("Item Type")
plt.ylabel("Total Sales")

for bar in bars:
    plt.text(bar.get_x()+bar.get_width()/2,bar.get_height(),
             f'{bar.get_height():,.1f}', ha='center',va='bottom',fontsize=8)
    
plt.tight_layout()
plt.show()
```

### **Total Sales Based On Outlet Location And Item Fat Content**

```python
Content_Outlet_Sales=df.groupby(['Outlet Location Type','Item Fat Content'])['Sales'].sum().unstack()
Content_Outlet_Sales=Content_Outlet_Sales[['Regular','Low Fat']]

ax = Content_Outlet_Sales.plot(kind='bar',figsize=(10,5))


plt.xlabel("Outlet Location Type")
plt.ylabel("Total Sales")
plt.legend(title='Item Fat Content')


plt.tight_layout()
plt.show()
```

### **Total Sales Based on Establishment year of outlet**

```python
Sales_by_year=df.groupby('Outlet Establishment Year')['Sales'].sum().sort_index()


plt.figure(figsize=(10,5))
plt.plot(Sales_by_year.index,Sales_by_year.values)


plt.title("Sales Based on Outlet Establishment Year")
plt.xlabel("Outlet Establishment Year")
plt.ylabel("Total Sales")


for x,y in zip(Sales_by_year.index,Sales_by_year.values):
    plt.text(x,y,f'{y:,.0f}',ha='center',va='bottom',fontsize=8)

plt.tight_layout()
plt.show()
```

### **Sales by Outlet Size**

```python
Sales_By_Outlet_Size=df.groupby('Outlet Size')['Sales'].sum()

plt.figure(figsize=(4,4))
plt.pie(Sales_By_Outlet_Size,
        labels=Sales_By_Outlet_Size.index,
         autopct='%1.2f%%',
          startangle=90)
plt.title("Sales based on the size of the outlet")
plt.tight_layout()
plt.show()
```

### **Sales by location**

```python
Sales_by_location=df.groupby('Outlet Location Type')['Sales'].sum().reset_index()

plt.figure(figsize=(10,4))
Sales_by_location=Sales_by_location.sort_values('Sales',ascending=False)

ax=sns.barplot(x='Sales',y='Outlet Location Type',data=Sales_by_location)


plt.title("Sales Based on Outlet Location")
plt.xlabel("Total Sales")
plt.ylabel("Outlet Location Type")

plt.tight_layout()
plt.show()
```
